# Transforma genero em binario,seleciona a coluna do csv,retira colunas não necessarias,transforma valores nulos em 0


```python
def Formatar(df):
    # transforma genero em binario
    df['Genero'] = df['Sex'].map({'female': 0, 'male': 1}).astype(int)
    # seleciona a coluna do csv
    ids = df['PassengerId'].values
    # retira colunas
    df = df.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'PassengerId',
                  'Embarked', 'SibSp', 'Parch', 'Fare'], axis=1)
    # transforma valores nulos em 0
    df = df.fillna(0)

    return df, ids
```

# Leitura de arquivo e conversao para array

```python
train = pd.read_csv('train.csv', header=0)
train.head()
```

```python
train, ids = Formatar(train)
test = pd.read_csv('test.csv', header=0)
test, ids = Formatar(test)
train_array = train.values
test_array = test.values
```
